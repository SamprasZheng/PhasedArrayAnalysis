```python
import numpy as np
import glob
import os
import matplotlib.pyplot as plt
dir = os.getcwd()
```

```python
phi=90;theta=-60
phase_h = np.array([[0  ,0  ,0  ,0  ,0  ,0  ,0  ,0  ],
                    [180,180,180,180,180,180,180,180],
                    [0  ,0  ,0  ,0  ,0  ,0  ,0  ,0  ],
                    [180,180,180,180,180,180,180,180],
                    [0  ,0  ,0  ,0  ,0  ,0  ,0  ,0  ],
                    [180,180,180,180,180,180,180,180],
                    [0  ,0  ,0  ,0  ,0  ,0  ,0  ,0  ],
                    [180,180,180,180,180,180,180,180]])
phase_v = np.array([[270,90,270,90,270,90,270,90],
                    [270,90,270,90,270,90,270,90],
                    [270,90,270,90,270,90,270,90],
                    [270,90,270,90,270,90,270,90],
                    [270,90,270,90,270,90,270,90],
                    [270,90,270,90,270,90,270,90],
                    [270,90,270,90,270,90,270,90],
                    [270,90,270,90,270,90,270,90]])
phase_shift= round(np.sin(theta*np.pi/180)*180,0)
if phi==0:
    for i in range(8):
        for j in range(8):
            # print(str(i)+","+str(j))
            phase_v[i][j] = (phase_v[i][j]+phase_shift*(i)) % 360
            phase_h[i][j] = (phase_h[i][j]+phase_shift*(i)) % 360
if phi==90:
    for i in range(8):
        for j in range(8):
            # print(str(i)+","+str(j))
            phase_v[j][i] = (phase_v[j][i]+phase_shift*(i)) % 360
            phase_h[j][i] = (phase_h[j][i]+phase_shift*(i)) % 360
    
```

```python
plt.imshow(phase_v, cmap='twilight_shifted', aspect='auto',vmin=0,vmax = 360, interpolation='nearest')
for i in range(8):
    for j in range(8):
        plt.annotate(f'{phase_v[i, j]:.0f}', xy=(j, i), ha='center', va='center', color='white')

# plt.axis('off')
plt.xlabel('x-elements')
plt.ylabel('y-elements')

colorbar= plt.colorbar()
plt.show
```
